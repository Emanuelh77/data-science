# Matplotlib!!

```python
import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
population = [2.519, 3.692, 5.263, 6.972] #in billions
plt.plot(year, population, 'g') #'g' for green
plt.xlabel('year')
plt.ylabel('population')
plt.title('# of people in billions per year')
plt.show()
plt.clf() #cleans up the graph to show another
plt.hist(population, bins = 3) #histogram, default bins is 10
plt.show()
```

the "plt.xscale('log')" allows you to display logarithmic scale for when numbers take up too much space. So instead of having 10,000,000,000,000 you'd have 10^12
